```python
#Gopalam. Rahul Malik
#AM.EN.U4EEE20030
import numpy as np
import pandas as pd

data=pd.read_csv("iris_dataset.csv",skiprows=1,header=None)
data.head(10)    

# We are splitting data to train and test.  

# Till or except -1 we taking Rest as x values

X=data.iloc[:,:-1].values
Y1=data.iloc[:,-1].values
Y=Y1.reshape(-1,1)
from  sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=41)

#Need to specify or else it take default value.

from sklearn import tree

#We have different classifiers inside this function like tree.decision tree classifiers.

classifier=tree.DecisionTreeClassifier(min_samples_split=3,max_depth=3,criterion="entropy")

#In order to limit 3 layers.

# Method of approach to find the decision value.

classifier.fit(X_train,Y_train)

#Used to training the model
classifier.score(X_test,Y_test)
tree.plot_tree(classifier)

#plotting decision tree
```
